```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def p2f(x):
    return float(x.strip('%'))/100

df = pd.read_csv("https://raw.githubusercontent.com/visualization-project-group/surprise-map-assignment/master/hate_crime.csv",usecols=['STATE_NAME', 'BIAS_DESC', 'DATA_YEAR'])
df3 = df.query("DATA_YEAR >= 2015 & DATA_YEAR <= 2018").groupby(['STATE_NAME']).agg('count')
df3 = df3.drop(columns = ['DATA_YEAR'])
df3 = df3.rename(columns = {'BIAS_DESC': 'TOTAL_COUNT_1518'})


df4 = df.query("DATA_YEAR >= 2015 & DATA_YEAR <= 2018 & BIAS_DESC == 'Anti-Asian'").groupby(["STATE_NAME"]).count()
df4 = df4.drop(columns = ['DATA_YEAR'])
df4 = df4.rename(columns = {'BIAS_DESC': 'TOTAL_ASIAN_COUNT_1518'})
df5 = pd.merge(df3,df4,on='STATE_NAME', how='inner')
df5['ASIAN_RATE_1518'] = df5['TOTAL_ASIAN_COUNT_1518']/df5['TOTAL_COUNT_1518']
df5.reset_index(inplace=True)
df5 = df5.rename(columns = {'STATE_NAME': 'State'})
df_twt = pd.read_csv("https://raw.githubusercontent.com/visualization-project-group/surprise-map-assignment/master/tweeter_data.csv")

df_comb = pd.merge(df5,df_twt, on = 'State', how='inner')
df_comb
```
